* **언어분야** : chatbot

*언어분야* 에서는 Chatbot을 생각했습니다.

__chatbot__ 이란 : 문자 또는 음성으로 대화하는 기능이 있는 컴퓨터 프로그램 또는 인공 지능 프로그램을 의미한다.

챗봇은 소셜네트워크, 병원, 쇼핑몰 외 여러 가지 Q&A 등의 고객서비스나 정보수집 용도로 다양하게 활용되고 있습니다.  
인공지능의 핵심적인 이론이라 할 수 있는 딥러닝을 적용한 가장 대표적인 사례라고 할 수 있습니다. 

챗봇은 과거에는 간단히 입력받은 단어에 대응하는 단어를 출력하는 간단한 방식 이였습니다. 이는 시간적인 순서에 큰 영향을 받고, 문맥을 파악하기 어렵다는 단점이 있습니다. 
이러한 문제를 보안하고 해결하기 위해서는 단어와 단어 사이의 관계를 파악하는 것을 넘어서 문장의 의도와 개체를 파악하는 것이 중요했고 기술의 발전을 통해 단점들을 하나하나 해결해가고 있습니다.

현재 챗봇에서 가장 많이 사용하는 방법인 __Char-CNN__은 CNN(convolutional Neural Network)방식을 사용합니다.
CNN은 이미지, 단어를 분석하기 위해 패턴을 찾는데 유용한 알고리즘입니다. 데이터에서 이미지나 단어, 문장을 보고 직접 학습하고 패턴을 사용하여 분류합니다.
질문-답변 데이터만 충분히 확보가 된다면 딥러닝 방식을 통해서 더욱 높은 결과 값을 가지는 챗봇을 개발할 수 있을 것입니다.

* **음성분야** : siri, 휘센 인공지능 에어컨, 인공지능 세탁기

*음성분야* 에서 siri, 휘센 인공지능 에어컨, 인공지능 세탁기를 생각하게 된 이유는
실제로 사용 중이고, 인공지능 에어컨 음성인식을 위해 필요한 데이터를 수집하는 프로그램에 실제로 참여해보았기 때문입니다. 
지금 사용하고 있는 핸드폰 아이폰에서는 siri라는 인간의 음성을 듣고 인식하여 간단한 조작을 자동으로 해주는 인공지능 플랫폼이 탑재되어 있습니다. siri야 1시간 뒤에 알람 맞춰줘 라는 간단한 대화만으로도 여러 번 터치해야하는 불편함을 감소시켜 주고 있습니다.

휘센 스마트 에어컨은 실제로 스마트 에어컨의 음성인식 서비스를 위한 데이터를 모으는 프로그램에 참여해 보았습니다. 
녹음실에서 백여 가지 넘는 문장들을 전라도 사투리로 녹음하는 프로그램 이였고, 저 뿐만 아니라 많은 사람들의 목소리를 녹음하였습니다. 

음성분야의 siri, 휘센 인공지능 에어컨은 음성 인식, 자연어 처리 기술이 탑재되어 있는 인공지능 플랫폼이 딥러닝 기술을 기반으로 사용자의 언어 사용 패턴까지도 스스로 학습하여 결과적으로 고객의 언어를 입력받아 처리할 수 있다고 합니다. 

아직까지는 정확한 작동원리에 대해서는 알지 못하지만, 수업과 과제를 통해서 자세히 알아가고 싶습니다. 
현재 가장 제가 가장 자주 사용하는 분야이고 많은 사람들이 이용하는 분야이기 때문에 더욱 정교하고 신뢰성 높은 기술로 발전된다면 더욱 편리한 생활을 만들어 줄 수 있는 분야라고 생각합니다. 

* **이미지분야** : 구글 이미지 검색

*이미지분야* 에서는 한번쯤은 이용해 보았을 법한 구글의 이미지 검색 기능을 생각하였습니다.

이미지 분야도 CNN방식을 이용합니다. 입력받은 이미지를 행렬로 표현하여 처리합니다. CNN은 합성곱층과 풀링층으로 나누어져 있습니다. 합성곱층에서는 간단히 이미지의 특징을 추출하고 풀링층에서는 특징을 강화하고 이미지의 크기를 축소합니다. 이 과정을 반복하여 처리된 결과를 출력하는 방식입니다. 

* **자율주행분야** : 자율주행자동차

*자율주행분야* 에서는 최근 가장 이슈가 되고, 차세대자동차 기술 중 빼놓을 수 없는 기술중 하나입니다. 

닛산, 현대, 아우디, bmw와 같은 여러 자동차회사에서 자율주행 자동차 서비스를 실현 시키기 위해 노력하고 있습니다.

IT기술이 발전됨에 따라 자동차의 첨단화, 지능화가 가속화 되고 있습니다. 자율주행자동차는 주행환경을 인식하여 운전을 보조하거나 스스로 주행하는 기술을 지닌 차세대 지능형 자동차라고 할 수 있습니다.

주행환경 인식기술을 바탕으로 하여 자율주행이 진행되는데 GPS, 카메라, Laser radar센서 등 다양한 센서를 통해 차량, 차선, 장애물 등을 인식하여 주행합니다. 이러한 주행환경 인식 및 자율주행 기술을 안정적으로 제어하기 위해 제어알고리즘 설계, 시뮬레이션, 시험을 통한 제어알고리즘 검증 등의 일련의 과정이 필요하고 이러한 과정들을 통해 제어시스템의 성능향상 및 신뢰성을 확보합니다.

자율주행 시스템의 소프트웨어는 센서 인터페이스, 인식부, 제어부, 차량 인터페이스로 구성됩니다. 각종 센서들로부터 데이터를 취득하고 인식부에서 이용할 수 있는 정보로 가공하게 됩니다. *인식부* 에서는 취득된 데이터를 바탕으로 차량의 위치, 장애물, 도로경계, 차선 등의 자율주행에 필요한 정보로 변환하고 취득된 정보를 융합하여 인식의 정확도 및 신뢰도를 높이게 됩니다. *제어부* 에서는 인식부에서 생성된 경로를 정확하게 추종할 수 있도록 차량의 거동을 제어하게 됩니다. 

자율주행기술은 여러 가지 복합된 분야의 집합체라고 생각합니다. 차량의 위치정보 추정부터 시작하여 장애물 및 도로 경계 인식, 차선 및 횡단보도 인식, 도착지까지의 경로 생성 및 각 도로 상황에 맞는 적절한 주행법 설정까지 다양한 분야의 전문지식이 필요합니다. 

인공지능 관점에서 보았을 때 각 센서들의 정보를 취득하고 취득된 정보를 다중모델필터를 통하여 각각의 상황에 맞는 정확한 결과 값을 도출해 내는 제어알고리즘 설계하여야 합니다. 

일상생활에서부터 미래를 위한 설계를 위해 인공지능 기술은 빼놓을 수 없는 분야입니다.
이미 많은 생활 속 인공지능 기술이 나왔고, 이러한 사회적 흐름과 시대적 흐름에서 인공지능 기술을 습득한다는 것은 매우 중요하고 가치 있다고 생각합니다. 사람이 살아가는 모든 생활 속에 인공지능 기술을 접목할 수 있게 되는 날이 머지않았다고 생각하고 그 안에서 함께 유용한 인공지능 기술들을 만들고 개발하고 싶습니다.
